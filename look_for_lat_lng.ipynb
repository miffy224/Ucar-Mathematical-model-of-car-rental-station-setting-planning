```python
import requests
import urllib.parse
import json
import csv
import pandas as pd
import numpy as np

Site_info = pd.read_csv('site.csv');

def get_latitude_longtitude(address):
    # decode url
    address = urllib.parse.quote(address)
    url = "https://maps.googleapis.com/maps/api/geocode/json?address=" + address +"&key=AIzaSyAEqF3zIvchAyTuRlReopab6eaD1JGveV0"
    while True:
        res = requests.get(url)
        js = json.loads(res.text)

        if js["status"] != "OVER_QUERY_LIMIT":
            break

    result = js["results"][0]["geometry"]["location"]
    lat = result["lat"]
    lng = result["lng"]

    return lat, lng

```

```python
for i in range(0,152):
    address = Site_info['地點'][i]
    lat, lng = get_latitude_longtitude(address)
    print(lat,",", lng)
    
```
